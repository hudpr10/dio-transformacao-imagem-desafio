```python
# Para este algoritmo as seguintes bibliotecas foram utilizadas:
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# Utiliza a biblioteca do MatPlotLib para exibir a imagem
imagem = mpimg.imread("./img/gato.jpg")

plt.imshow(imagem)
plt.axis("off")
plt.show()

print("1. Pixel qualquer:", imagem[700, 600])
print("2. Pixel qualquer:", imagem[1, 100])
print("3. Pixel qualquer:", imagem[1000, 300])
```

```python
# Faz uma cpia da imagem original
imagem_tons_cinza = imagem.copy()

# Para criar o efeito de tons de cinza, os pixels precisam ter valores prximos em todos os canais (exemplo R: 150, G: 150 e B: 150)
# Esse trecho de cdigo faz exatamente isso, ele utiliza os valores do canal vermelho para aplicar nos canais verde e azul, criando o efeito de tons de cinza
imagem_tons_cinza[:, :, [1, 2]] = imagem_tons_cinza[:, :, [0, 0]]

plt.imshow(imagem_tons_cinza)
plt.axis("off")
plt.show()

print("1. Pixel qualquer:", imagem_tons_cinza[700, 600])
print("2. Pixel qualquer:", imagem_tons_cinza[1, 100])
print("3. Pixel qualquer:", imagem_tons_cinza[1000, 300])
```

```python
# Faz uma cpia da imagem em tons de cinza
imagem_preto_e_branco = imagem_tons_cinza.copy()

# Para transformar a imagem em preto e branco foi usada a seguinte lgica:
# 1. A imagem a ser manipulada deveria ser transformada em tons de cinza, assim como feito no mdulo anterior
# 2. Para cada pixel da imagem uma condio deve ser testada, se o valor do pixel for maior que 140, ele dever receber o novo valor de 255 (branco total)
# 2.1. Se o valor do pixel for menor que 140, ele dever receber o novo valor de 0 (preto total)
imagem_preto_e_branco[imagem_preto_e_branco[:, :, 0] > 140] = [255, 255, 255]
imagem_preto_e_branco[imagem_preto_e_branco[:, :, 0] < 140] = [0, 0, 0]

plt.imshow(imagem_preto_e_branco)
plt.axis("off")
plt.show()

print("1. Pixel qualquer:", imagem_preto_e_branco[700, 600])
print("2. Pixel qualquer:", imagem_preto_e_branco[1, 100])
print("3. Pixel qualquer:", imagem_preto_e_branco[1000, 300])
```
